```python
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images
```

```python
images = load_images_from_folder('./data/train/')
```

```python
names = os.listdir('./data/train/')
print(len(names))
```

```python
assign = []
for i in range(len(names)):
    name = names[i].split('.')
    if name[0] == 'dog':
        assign.append(0)
    else:
        assign.append(1)

assign = np.array(assign).reshape((-1,1))
print(assign.shape)
print(assign[5],assign[1],assign[2],sep=",")
```

```python
train_x = np.array(images)
print(train_x.shape)
train_x = train_x.reshape((-1,1))
print(train_x.shape)
train_y = np.array(assign)
print(train_y.shape)

plt.imshow(train_x[0][0])
print(train_y[0])
```
